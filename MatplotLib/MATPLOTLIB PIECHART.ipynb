#  PIECHART

```python
import matplotlib.pyplot as plt
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students, labels = classes)
plt.show()
```

```python
#explode
#slicing part of pie chart
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode)
plt.show()
```

```python
#color
color=['c','b','r','y','g']
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode , colors=color)
plt.show()
```

```python
#autopercentage

#values in percentage format
#o.1 indictes single value after . 
color=['c','b','r','y','g']
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode , colors=color , autopct="%0.1f%%")
plt.show()
```

```python
#shadow parameter
#boolean valyue
color=['c','b','r','y','g']
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode , colors=color , autopct="%0.1f%%", shadow= True)
plt.show()
```

```python
#radius
color=['c','b','r','y','g']
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode , colors=color , 
        autopct="%0.1f%%", shadow= True, radius= 1.8)
plt.show()
```

```python
#startangle 
explode=[0.04,0.,0.1,0,0]
plt.pie(class1_students, labels = classes, explode = explode , colors=color , 
        autopct="%0.1f%%", shadow= True, radius= 1.8 , startangle=90)
plt.show()
```

```python
#textprop
explode=[0.04,0.,0.1,0,0]
textprop={"fontsize": 15,}
plt.pie(class1_students, labels = classes, explode = explode , colors=color , 
        autopct="%0.1f%%", shadow= True,radius= 1.5 , textprops= textprop)


plt.show()
```
