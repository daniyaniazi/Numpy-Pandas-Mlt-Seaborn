# Histograms

```python
#importing libs
import matplotlib.pyplot as plt
import numpy as np
import random 
```

```python
# Random Data
ml_stu_age=np.random.randint(18,45,(100))
py_stu_age=np.random.randint(15,44,(100))
```

```python
plt.hist(ml_stu_age)
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8)
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="step")
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="bar", orientation ="horizontal")
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="bar" , color="r")
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="bar" , color="r" , label="ml student")
plt.legend()
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="bar" , color="r" , label="ml student")
plt.hist(py_stu_age , bins ,rwidth= 0.8, histtype ="bar" ,orientation ="horizontal", color="b" , label="py student")

plt.legend()
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_stu_age , bins ,rwidth= 0.8, histtype ="bar" , color="r" , label="ml student")
plt.hist(py_stu_age , bins ,rwidth= 0.8, histtype ="bar" , color="b" , label="py student")

plt.legend()
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
#  Fig size
plt.figure(figsize=(16,9))
plt.hist([ml_stu_age,py_stu_age] , bins ,rwidth= 0.8, histtype ="bar" , color=["r","b"] , label=["ml student","py student"])

plt.legend()
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
#  Fig size
plt.figure(figsize=(16,9))
plt.hist([ml_stu_age,py_stu_age] , bins ,rwidth= 0.8, histtype ="bar" , color=["r","b"] , label=["ml student","py student"])

plt.legend()
plt.title("ML STUDENTS HISTOGRAM")
plt.xlabel("students age category" )
plt.ylabel("no of stu age" )
plt.show()
```
