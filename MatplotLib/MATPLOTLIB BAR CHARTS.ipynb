# Bar Charts

```python
import matplotlib.pyplot as plt
import numpy as np
import random 
```

```python
classes=["Python", 'R',' AI','ML','DS']
class1_student=[30,10,20,25,10]
class3_student=[40,5,20,20,10]
class2_student=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_student)
```

```python
plt.barh(classes,class1_student)
```

```python
plt.bar(classes,class1_student,width=0.2, align="edge", color="red", edgecolor="y", linewidth=5,alpha=0.8 , linestyle= "--", label="class 1 students")
plt.legend()
```

```python
from matplotlib import style
plt.figure(figsize=(16,9))
style.use('ggplot')
# barh dont take width rgument remember
plt.bar(classes,class1_student,width=0.2, align="edge", color="red", edgecolor="y", linewidth=2,alpha=0.8 ,
        linestyle= "--", label="class 1 students")
plt.legend()
plt.title("BAR CHART", fontsize=25)
plt.xlabel("x" )
plt.ylabel("y" )
plt.show()
```

```python
plt.figure(figsize=(12,9))
style.use('ggplot')
# barh dont take width rgument remember
plt.bar(classes,class1_student,width=0.2, color="red",label="class 1 students")
plt.bar(classes,class2_student,width=0.2, color="blue",label="class 2 students")
plt.legend()
plt.title("BAR CHART", fontsize=25)
plt.xlabel("x" )
plt.ylabel("y" )
plt.show()
```

##### seperate bar charts

```python
classes_index= np.arange(len(classes))
plt.figure(figsize=(12,9))
width=0.2 
plt.bar(classes_index,class1_student,width, color="red",label="class 1 students")
plt.bar(classes_index+width,class2_student,width, color="blue",label="class 2 students")
plt.bar(classes_index+width+width ,class3_student,width, color="y",label="class 3 students")
plt.legend()

plt.xticks=(classes_index+width,classes)
plt.title("BAR CHART", fontsize=25)
plt.xlabel("x" )#hbar ylabel
plt.ylabel("y" )#hbar xlabel
plt.show()
```
