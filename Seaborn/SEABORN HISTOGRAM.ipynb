```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

# SEABORN DISPLOT

```python
tips_df=pd.read_csv("tipscsv.csv")
```

```python
tips_df
```

```python
#numeric data
sns.distplot(tips_df["size"])
```

```python
"""  a,
    bins=None,
    hist=True,
    kde=True,
    rug=False,
    fit=None,
    hist_kws=None,
    kde_kws=None,
    rug_kws=None,
    fit_kws=None,
    color=None,
    vertical=False,
    norm_hist=False,
    axlabel=None,
    label=None,
    ax=None"""
```

```python
sns.distplot(tips_df["total_bill"],bins=100)
```

```python
sns.distplot(tips_df["total_bill"],bins=100 ,hist=False) #show kernel density estimater
```

```python
sns.distplot(tips_df["total_bill"],kde=False)
```

```python
sns.distplot(tips_df["total_bill"],rug=True) 
#data on x axis is shown
```

```python
sns.distplot(tips_df["total_bill"], fit=norm)  #normalize line plot  you can hide kde
```

```python
sns.distplot(tips_df["total_bill"],color="red")
```

```python
sns.distplot(tips_df["total_bill"],vertical=True)
```

```python
sns.distplot(tips_df["total_bill"],norm_hist=True)
```

```python
sns.distplot(tips_df["total_bill"],axlabel="TOTAL BILL")
```

```python
sns.distplot(tips_df["total_bill"],label="BILLS")
plt.legend() #you can give title
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],label="BILLS")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,20,25,30,35,40,45,50,55]
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],label="BILLS",bins=bins)
plt.legend()
plt.xticks(bins)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"],hist_kws={ "color":"g","edgecolor":"blue","linewidth":3,"linestyle":"--"})
```

```python
sns.distplot(tips_df["total_bill"],hist_kws={ "color":"g","edgecolor":"blue","linewidth":3,"linestyle":"--"},
            kde_kws={ "color":"black","linewidth":3,"linestyle":"--"})


# same for rug
```

```python
sns.distplot(tips_df["total_bill"],hist_kws={ "color":"g","edgecolor":"blue","linewidth":3,"linestyle":"--"},
            kde_kws={ "color":"black","linewidth":3,"linestyle":"--"}, fit=norm,
            fit_kws={ "color":"white","linewidth":3,"linestyle":"--"})
```
