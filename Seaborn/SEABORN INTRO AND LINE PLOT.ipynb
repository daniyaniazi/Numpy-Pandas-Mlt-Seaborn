```python
tips_df= sns.load_dataset("tips")
tips_df
```

```python
tips_df.to_csv('tipscsv.csv', index = False)
```

```python
tips_df.shape
```

```python
sns.lineplot(x="total_bill", y="tip", data= tips_df)
```

```python
sns.lineplot(x="tip", y= "size", data= tips_df)
```

```python
sns.lineplot(x="size", y= "tip", data= tips_df , hue = "sex")#category value
```

```python
sns.lineplot(x="size", y= "tip", data= tips_df , hue = "sex", style= "sex")
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style= "sex", palette="Dark2_r")
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style= "sex", palette="Dark2_r" , dashes= False)
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style= "sex", palette="Dark2_r" , dashes= False , markers=["o","<"])
```

```python
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style= "sex", palette="Dark2_r" , dashes= False , legend="full")
```

```python
sns.set(style= "darkgrid")
sns.lineplot(x="size",y="tip",data=tips_df,hue="sex",style= "sex", palette="Dark2_r" , dashes= False)
```
