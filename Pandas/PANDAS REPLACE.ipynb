Values to the dataframe afre replaced with other values dinamicallly

```python
#dataframe.replace()
#parameter : to_replace, limit, regex, value , inplace ,method
```

```python
import pandas as pd
```

```python
df=pd.read_csv("Fortune_10.csv") #C:\\Users\\DANIA NIAZI\\Desktop\\DANIYA\\MACHINE LEARNING\\Fortune_10.csv
df
```

```python
df.replace(to_replace="Health", value="Finance")
```

```python
df.replace("Health","Finance") 
#short cut
```

#### replace no of value with a single values

```python
df.replace([1,2,3,4,5,6,7,8,9],"a")
```

##### replace no of value with no of  values

```python
df.replace([1,2,3,4,5,6,7,8,9],['a','b','c','d','e','f','g','h','i'])
```

### specific col value replace by sepcific value by using dictionary

```python
df.replace({"Industry":"Health"},"education")
#{col : value} ,"new value"
```

```python
#two or more columns 
```

```python
df.replace({"Industry":"Health","Name":"Ganzlax"},"education")
```

## regex

```python
df.replace("[A-Za_z]",0, regex=True)
#string pattern to int
```

### dictionary for specific column

```python
df.replace({"Industry":"[A-Za_z]"},0,regex=True)
```

## Method

```python
df.replace("Health", method="ffill") #value , fill
```

```python
df.replace("Health", method="bfill")
```

## limit


```python
df.replace("Stripfind", method="bfill", limit=1)
```

```python
df.replace("Stripfind", method="bfill", limit=2)
```

## inplace

```python
df.replace("no","maybe",inplace=True)
```
