FILLNA FILLS THE NAN VALUES WITH GIVEN  VALUES (INPUT) IN DATASET
SYNTAX: DataFrame.fillna()


```python
import pandas as pd
```

```python
df=pd.read_csv("Fortune_10.csv") #C:\\Users\\DANIA NIAZI\\Desktop\\DANIYA\\MACHINE LEARNING\\Fortune_10.csv
df
```

```python
#you can see the missing value showing nan
df.fillna(0)
```

```python
df.fillna(4)
#fill scalar value
```

### Fill specific columns by dictionaries

```python
df.fillna({"Name": "none", "Industry":"no"})
#{"col name" :"value to be displayed"}
```

### method parameter

```python
df.fillna(method="ffill") #forward fill
#method use when want to fill with previous or forward value
```

```python
df.fillna(method="pad")
```

```python
df.fillna(method="bfill")
```

### AXIS

```python
df.fillna(axis=0)
```

```python
#we need to identitify value 
df.fillna(method="ffill", axis=0)
```

```python
df.fillna(method="ffill", axis=1)
```

```python
df.fillna(method="bfill", axis=1)
```

### Limit parameter

```python
df.fillna(0, limit=1) #need to specify value
```

```python
#only 1 fill
```

```python
df.fillna(method="ffill", limit=2)
```

### inplace 

```python
#default vadflue false
#modify data frame on True
#don't cretae new data frame
```

```python
df.fillna(5,inplace=True)
df
```
